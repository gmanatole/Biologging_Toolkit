```python
import sys
import numpy as np
sys.path.append('Biologging_Toolkit/src/')
from Biologging_Toolkit.wrapper import Wrapper
from Biologging_Toolkit.processing.Dives import Dives
```

```python
depids = [ 'ml18_296a', 'ml18_294b','ml19_292a', 'ml19_292b', 'ml19_293a', 'ml19_294a', 'ml20_293a', 'ml20_296b', 'ml20_313a', 'ml21_295a', 'ml21_305b', 'ml17_280a']
```

```python
inst = Wrapper(depids[4], path = f'E:/individus_brut/individus/{depids[4]}/')
dive = Dives(depids[4], path = f'E:/individus_brut/individus/{depids[4]}/')
```

```python
for depid in depids[-1:] :
    dive_wind = []
    path = f'D:/individus_brut/individus/{depid}/'
    inst = Wrapper(depid, path = path)
    wind = inst.ds['cfosat'][:].data
    qual = inst.ds['cfosat_quality'][:].data
    dive_array = inst.ds['dives'][:].data
    dive = Dives(depid, path = path)
    for dive_number in dive.dive_ds.dive :
        _wind = wind[(dive_array == dive_number) & (qual == 1)]
        dive_wind.append(np.nanmean(_wind[_wind >= 0]))
    dive.dive_ds['cfosat_wind'] = dive_wind
    dive.dive_ds.to_csv(dive.dive_path, index = None)
    inst.ds.close()
```
